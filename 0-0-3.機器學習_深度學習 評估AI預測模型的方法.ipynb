## 1.混淆矩陣(confusion matrix)

## 3. k-fold Cross validation
### 交叉驗證技術叫做K折交叉驗證(K-fold Cross Validation)，將整個數據集拆分成訓練集和測試集，在訓練集上訓練模型，
### 在測試集上評估模型。
### 透過資料間的重複採樣過程，用於評估機器學習模型並驗證模型
### 在 K-Fold 的方法中我們會將資料切分為 K 等份，K 是由我們自由調控的，以下圖為例：假設我們設定 K=10，也就是將訓練集切割為十等份。
### 這意味著相同的模型要訓練十次，每一次的訓練都會從這十等份挑選其中九等份作為訓練資料，剩下一等份未參與訓練並作為驗證集。
### 因此訓練十回將會有十個不同驗證集的 Error，這個 Error 通常我們會稱作 loss 也就是模型評估方式。最終把這十次的 loss 加總起來取平均就可以當成最終結果。


### k-fold的k代表的是將訓練資料切分為k組，每次將(k-1)組資料作為訓練資料，1組作為驗證資料，重複迭代k次。K的數量可根據資料數量來做決定。


## 4.損失函數(loss function)
### 損失函數是一種評估 “你的算法 / 模型對你的數據集預估情況的好壞” 的方法。如果你的預測是完全錯誤的，你的損失函數將輸出一個更高的數字。如果預估的很好，它將輸出一個較低的數字。當調整算法以嘗試改進模型時，損失函數將能反應模型是否在改進。“損失”有助於我們瞭解預測值與實際值之間的差異。
### 常用損失函數:
### Mean Error (ME)

### Mean Squared Error (MSE)


### Mean Absolute Error (MAE)


### Root Mean Squared Error (RMSE)



### 交叉熵(Cross Entropy)
#### 最常使用於分類問題的損失函數(loss functions)。交叉熵(Cross Entropy)又有個別名log loss


## AI預測模型評估結果不佳的原因
### <li> 擬合不足與過度擬合(訓練不足與過度訓練)
### 擬合不足模型會出現較高的預測錯誤偏差，會為訓練資料集和測試集提供不正確的結果(訓練不足)。
### 過度擬合模型會出現較高的預測結果差異，會對訓練集提供準確的預測結果，但會對測試集提供不準確的預測結果。模型訓練越多(過度訓練)，導致的差異可能會增加。
### 所以在訓練模型時，要尋找擬合不足與過度擬合間的最佳位置(時機點)。

##  AI預測模型評估結果不佳的改善方法
### 若是擬合不足,就給模型有適當的時間長度來訓練模型並且改善資料。
### 若是過度擬合,可以透過提供更多樣化和擴展的訓練資料集，或者使用一些策略 (如下所示策略)，來防止過度擬合。
### 1.提前停止訓練
### 2.剪除若干特徵或參數，並排除不相關的特徵。
### 3.正規化資料，嘗試消除那些不會影響預測結果的因素。
### 4.整合: 將多個單獨的機器學習演算法(預測模型)相結合。有些模型被稱為弱學習者，因為其結果往往不準確。整合法結合所有弱學習者，以取得更準確的結果。
### 5.資料增強:資料增強是一種機器學習技術，每次模型處理範例資料時，都會對其稍做變更。例如，將平移、翻轉和旋轉等轉換套用至輸入影像。
